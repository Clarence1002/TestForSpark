```python
## Cleaning Up
```

```python
from pyspark.sql.functions import format_number, col

# Let's start off with this. Just grouping by job and presenting the mean.
group_job_df1 = df1.groupBy('quality').mean()
group_job_df1.show()
```

```python
# Now that we've calculated the mean, the values for blue-collar and technician are extremely long. 
# We can use format_number to reduce the total amount of decimals. 
# The number two represents the amount of decimals we want to be displayed.
group_job_df1 = group_job_df1.select('quality',
                                   format_number('avg(fixed acidity)',2),
                                   format_number('avg(volatile acidity)',2),
                                   format_number('avg(citric acid)',2),
                                   format_number('avg(residual sugar)',2),
                                   format_number('avg(chlorides)',2),
                                   format_number('avg(free sulfur dioxide)',2),
                                   format_number('avg(total sulfur dioxide)',2),
                                   format_number('avg(density)',2),
                                   format_number('avg(pH)',2),
                                   format_number('avg(sulphates)',2),
                                   format_number('avg(alcohol)',2))
group_job_df1.show()
```

```python
# But now the column names look quite unprofessional. We can assign an alias to rename each of them.
group_job_df1 = group_job_df1.select(col('quality').alias('Quality Category'),
                                   col('format_number(avg(density), 2)').alias('Average density'),
                                   col('format_number(avg(pH), 2)').alias('Average pH'),
                                   col('format_number(avg(sulphates), 2)').alias('Average sulphates'),
                                   col('format_number(avg(alcohol), 2)').alias('Average alcohol'))
group_job_df1.show()

# Finally, let's sort the DataFrame by age.
group_job_df1 = group_job_df1.orderBy('Average alcohol')

print('Average Age, Balance and Duration by Job Category')
group_job_df1.show()
```
